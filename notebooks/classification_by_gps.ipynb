```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Example data: list of tuples with (latitude, longitude)
gps_coordinates = [
    (34.052235, -118.243683),
    (34.052331, -118.243127),
    (34.049908, -118.243159),
    # Add more coordinates here
]

# Convert the list of tuples to a numpy array
gps_coordinates_np = np.array(gps_coordinates)

# Create a KMeans instance with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)

# Fit the KMeans algorithm to the GPS data
kmeans.fit(gps_coordinates_np)

# Get the cluster assignments
labels = kmeans.labels_

# Print the assigned groups
for idx, label in enumerate(labels):
    print(f"Child {idx} is in group {label}")

# Plotting the clusters
plt.scatter(gps_coordinates_np[:, 0], gps_coordinates_np[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Clustering of Children\'s Homes')
plt.show()
```
